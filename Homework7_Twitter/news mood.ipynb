```python
import tweepy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import numpy 
import json
import requests
import random
import numpy as np
import time

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

output_data_file1 = "output_data/fox.csv"
output_data_file2 = "output_data/nytimes.csv"
output_data_file3 = "output_data/cbs.csv"
output_data_file4 = "output_data/bbc.csv"
output_data_file5 = "output_data/cnn.csv"
output_data_file6 = "output_data/combined.csv"
```

```python
# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target1 "@foxnews"
target1 = "@foxnews"
number = 1
sentiments1 = []

for x in range(5):

    # Get all tweets from home feed
    public_tweets = api.user_timeline(target1)
    
    
# Loop through all tweets
for tweet in public_tweets:
    # print(json.dumps(tweet, sort_keys=True, indent=4))

    # Run Vader Analysis on each tweet
    compound = analyzer.polarity_scores(tweet["text"])["compound"]
    pos = analyzer.polarity_scores(tweet["text"])["pos"]
    neu = analyzer.polarity_scores(tweet["text"])["neu"]
    neg = analyzer.polarity_scores(tweet["text"])["neg"]

   
    name = "Fox"
    tweets_ago = number
    
    # Add sentiments for each tweet into an array
    sentiments1.append({"Media": name, 
                           "Date": tweet["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": number})
    number= number+1

# Convert sentiments to DataFrame
sentiments1_pd = pd.DataFrame.from_dict(sentiments1)
sentiments1_pd.head()

sentiments1_pd.to_csv(output_data_file1, index_label="Sentiments_ID")
```

```python
plt.figure(figsize=(9,6))

sns.set()



bbc= plt.scatter(combined_sentiments_pd[combined_sentiments_pd["Media"]=="BBC"]["Tweets Ago"], 
                 combined_sentiments_pd[combined_sentiments_pd["Media"]=="BBC"]["Compound"], 
                      s = 100,alpha = 1, marker="o",label = "BBC",edgecolors="black")

fox= plt.scatter(combined_sentiments_pd[combined_sentiments_pd["Media"]=="Fox"]["Tweets Ago"], 
                 combined_sentiments_pd[combined_sentiments_pd["Media"]=="Fox"]["Compound"], 
                      s = 100,alpha = 1, marker="o",label = "Fox",edgecolors="black")

cnn= plt.scatter(combined_sentiments_pd[combined_sentiments_pd["Media"]=="CNN"]["Tweets Ago"], 
                 combined_sentiments_pd[combined_sentiments_pd["Media"]=="CNN"]["Compound"], 
                      s = 100,alpha = 1, marker="o",label = "CNN",edgecolors="black")

nyt= plt.scatter(combined_sentiments_pd[combined_sentiments_pd["Media"]=="NewYork Times"]["Tweets Ago"], 
                 combined_sentiments_pd[combined_sentiments_pd["Media"]=="NewYork Times"]["Compound"], 
                      s = 100,alpha = 1, marker="o",label = "NYTimes",edgecolors="black")

cbs= plt.scatter(combined_sentiments_pd[combined_sentiments_pd["Media"]=="CBS"]["Tweets Ago"], 
                 combined_sentiments_pd[combined_sentiments_pd["Media"]=="CBS"]["Compound"], 
                      s = 100,alpha = 1, marker="o",label = "CBS",edgecolors="black")

lgnd = plt.legend(handles= [bbc, fox, cbs, nyt, cnn], title= 'Media Sources',loc=(1.1, 0.5))

plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")

plt.title("Sentiment Analysis of Media Tweets for (%s)" % (time.strftime("%x")))
plt.ylim(-1, 1)

plt.xlim(20, 0)

plt.savefig("output_data/Sentiment Analysis")

plt.show()
```
